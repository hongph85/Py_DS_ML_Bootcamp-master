```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
sns.barplot(data=tips, x='sex',y='total_bill')
```

```python
sns.barplot(data=tips, x='time',y='total_bill')
```

```python
import numpy as np
```

```python
sns.barplot(data=tips, x='sex',y='total_bill',estimator=np.std)
```

```python
sns.countplot(data=tips, y='sex')
```

```python
sns.boxplot(x='day', y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot( data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='day', y='total_bill',data=tips,palette='rainbow',hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill',data=tips,palette='rainbow',hue='smoker')
```
